```python
class Automovil:
    def __init__(self, marca, place, precio):
        self.marca = marca
        self.place = place
        self.precio = precio

    def leerdatos(self):
        self.marca = input("Ingrese la marca: ")
        self.place = input("Ingrese el lugar: ")
        self.precio = input("Ingrese el precio: ")
        return self.marca, self.place, self.precio

    def mostrardatos(self):
        print("Marca: ", self.marca)
        print("Lugar: ", self.place)
        print("Precio: ", self.precio)

# Fin de la clase automovil

# Creando los objetos carro1, carro2, carro3
print("Carro Numero 1")
carro1 = Automovil("", "", "") # Creando el objeto carro1
carro1.leerdatos()
carro1.mostrardatos()
print("--------------------------")
print("Carro Numero 2")
carro2 = Automovil("", "", "") # Creando el objeto carro2
carro2.leerdatos()
carro2.mostrardatos()
print("--------------------------")
print("Carro Numero 3")
carro3 = Automovil("", "", "") # Creando el objeto carro3
carro3.leerdatos()
carro3.mostrardatos()
print("--------------------------")
print("Software desarrollado por Meredith Aguirre NL = 2")
```
